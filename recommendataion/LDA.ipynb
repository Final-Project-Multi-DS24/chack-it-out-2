```python
# LDA
from sklearn.decomposition import LatentDirichletAllocation

td_lda = LatentDirichletAllocation(
    n_components=16, # n_components : 생성할 토픽의 개수
    random_state=42
)

td_lda.fit(cnt_feature_vector)
```

```python
# 토픽별 단어 추출
for topic_index, topic_score in enumerate(td_lda.components_):
  print(f"Topic # {topic_index}")

  # 단어의 인덱스 구하기. 점수가 높은 순서대로 구하기
  topic_word_indexes = topic_score.argsort()[::-1]
  top_indexes = topic_word_indexes[:10]

  author_feature_names = cnt_vectorizer.get_feature_names_out()

  top_features = [str(author_feature_names[i]) for i in top_indexes]
```
