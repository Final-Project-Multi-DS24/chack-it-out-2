# 추천 성능 향상
- 초기(step1) 모든 도서정보를 한번에 벡터화한 결과, 랜덤추천과 비슷한 추천 성능을 보임
- BoW를 확인한 결과 도서를 표현하는 단어의 수가 너무 적다고 판단하여 단어 수를 늘리기 위한 방안으로 각 도서 정보를 concat하는 방안을 고려함
- step2,3의 차이는 두 종류 서평의 합침 여부인데, 서평의 내용이 비슷하기 때문에 한꺼번에 고려해도 될 것이라고 판단함
- 최종적으로 step3(두 종류의 서평을 합쳐서 벡터화)한 경우 추천 성능이 가장 우수했음

#TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_vec = TfidfVectorizer(
    ngram_range=(1, 3),
    min_df = 0.07,
    max_df = 0.95
)
```

## step1 - 제목 + 목차 + 서평1 + 서평2

```python
# 전체 벡터화
total_list_sent = [' '.join(sent) for sent in total_mecab_list]
tf_total_vector = tf_vec.fit_transform(total_list_sent)
tf_total_matrix = tf_total_vector.toarray()
```

## step2 - concat(제목 + 목차 + 서평1 + 서평2)

```python
# 제목 벡터화
title_list_sent = [' '.join(sent) for sent in title_mecab_list]
tf_title_vector = tf_vec.fit_transform(title_list_sent)
tf_title_matrix = tf_title_vector.toarray()
```

```python
# 목차 벡터화
toc_list_sent = [' '.join(sent) for sent in toc_mecab_list]
tf_toc_vector = tf_vec.fit_transform(toc_list_sent)
tf_toc_matrix = tf_toc_vector.toarray()
```

```python
# full1 벡터화
full1_list_sent = [' '.join(sent) for sent in full_mecab_list]
tf_full1_vector = tf_vec.fit_transform(full1_list_sent)
tf_full1_matrix = tf_full1_vector.toarray()
```

```python
# full2 벡터화
full2_list_sent = [' '.join(sent) for sent in full2_mecab_list]
tf_full2_vector = tf_vec.fit_transform(full2_list_sent)
tf_full2_matrix = tf_full2_vector.toarray()
```

```python
import numpy as np
con_4f_matrix = np.concatenate((tf_title_matrix, tf_toc_matrix, tf_full1_matrix, tf_full2_matrix), axis=1)
```

## step3 - concat[제목 + 목차 + (서평1 + 서평2)]

```python
# 서평1 + 서평2 벡터화
tot_full_list_sent = [' '.join(sent) for sent in tot_full_mecab_list]
tf_tot_full_vector = tf_vec.fit_transform(tot_full_list_sent)
tf_tot_full_matrix = tf_tot_full_vector.toarray()
```

```python
con_3f_matrix = np.concatenate((tf_title_matrix, tf_toc_matrix, tf_tot_full_matrix), axis=1)
con_3f_matrix.shape
```

#추천

```python
from sklearn.metrics.pairwise import cosine_similarity
book_idx = 5108
cos_3f_sim = cosine_similarity([con_3f_matrix[book_idx]], con_3f_matrix)
cos_3f_sim
```

```python
sim_3f_scores = list(enumerate(cos_3f_sim[-1]))
sim_3f_scores = sorted(sim_3f_scores, key=lambda x : x[1], reverse=True)
sim_3f_scores[:10]
```

```python
bookidx_3f = [i[0] for i in sim_3f_scores]
df.iloc[bookidx_3f[:10], :]
```
