### fulldescription 공백 처리

```python
# fulldescription 없는 도서
lda_df[lda_df['fulldescription'].isnull() == True]
```

fulldescroption 컬럼이 결측값인 행을 description 내용으로 채우고 나서 채워진 description 내용이 결측값인 경우 toc컬럼의 내용으로 채우기
