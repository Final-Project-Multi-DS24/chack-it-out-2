```python
import pickle
with open("./Noun_mecab.pkl","rb") as a:
    token = pickle.load(a) 
```

```python
import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings(action='ignore')
from tqdm import tqdm 
from gensim import corpora 
import gensim 
import pyLDAvis.gensim_models 
from collections import Counter 

import urllib.request
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
from konlpy.tag import * 
from tqdm import tqdm

detokenized_doc = []
for i in range(len(token)):
    t = ' '.join(token[i])
    detokenized_doc.append(t)    
```

```python
    
# 토픽 확인 함수
def get_topics(components, feature_names, n=1000):
    features_li = []
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1))
        print([(feature_names[i], topic[i].round(2)) for i in topic.argsort()[:-n - 1:-1]])
        
        features = []
        for i in topic.argsort()[:-n - 1:-1]:
            feature = feature_names[i]
            features.append(feature)
            
        features_li.append(features)
    return features_li
        
    

# 상위 1,000개의 단어를 보존 
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(detokenized_doc)

lda_model = LatentDirichletAllocation(n_components=10,learning_method='online',random_state=777,max_iter=1)
lda_top = lda_model.fit_transform(X)

terms = list(vectorizer.vocabulary_.keys())

topics = get_topics(lda_model.components_,terms) 
```

시각화를 위한

```python
from gensim import corpora
dictionary = corpora.Dictionary(token)
corpus = [dictionary.doc2bow(text) for text in token]
print(corpus[1])
```

```python
import gensim
NUM_TOPICS = 10 
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
topics = ldamodel.print_topics(num_words=20)
for topic in topics:
    print(topic)
```

```python
import pyLDAvis
import pyLDAvis.gensim_models

pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(ldamodel, corpus, dictionary)
pyLDAvis.display(vis)
```
